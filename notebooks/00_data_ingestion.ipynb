# 00 – Data Ingestion
This part sets up Snowflake and brings the raw CSV files into the RAW layer. Just loading the data and making sure everything landed properly.


```
USE DATABASE AMAZON_REVIEWS_DB;
USE SCHEMA RAW;
USE WAREHOUSE COMPUTE_WH;
```

### Creating a stage
Setting up an internal Snowflake stage where the CSV files will be uploaded.


```
CREATE OR REPLACE STAGE AMAZON_STAGE
FILE_FORMAT = (TYPE = CSV FIELD_OPTIONALLY_ENCLOSED_BY='"' SKIP_HEADER=1);
```

### Checking the files in the stage
Just making sure all CSV files are actually there before loading them.


```
LIST @AMAZON_STAGE;
```

### Raw table schema
Building the table structure that matches the columns in the original dataset.


```
CREATE OR REPLACE TABLE REVIEWS_RAW (
    ID NUMBER,
    PRODUCTID VARCHAR,
    USERID VARCHAR,
    PROFILENAME VARCHAR,
    HELPFULNESSNUMERATOR NUMBER,
    HELPFULNESSDENOMINATOR NUMBER,
    SCORE NUMBER,
    TIME NUMBER,
    SUMMARY VARCHAR,
    TEXT VARCHAR
);
```

### Loading the raw data
Copying the CSV data from the stage into the RAW table. Allowing partial load in case a few rows are messy.


```
COPY INTO REVIEWS_RAW
FROM @AMAZON_STAGE
FILE_FORMAT = (FORMAT_NAME = 'AMAZON_CSV_FORMAT')
ON_ERROR = 'CONTINUE';
```

```
CREATE OR REPLACE FILE FORMAT AMAZON_CSV_FORMAT
TYPE = 'CSV'
FIELD_OPTIONALLY_ENCLOSED_BY = '"'
SKIP_HEADER = 1;
```

### Quick data check
Checking row counts and previewing a few rows to confirm everything loaded correctly.


```
SELECT COUNT(*) AS ROW_COUNT
FROM REVIEWS_RAW;

SELECT * FROM REVIEWS_RAW LIMIT 20;
```
